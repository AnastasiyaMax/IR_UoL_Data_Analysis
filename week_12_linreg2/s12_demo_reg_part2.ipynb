We will use the dataset on bike-sharing demand

*   **Date** - dd/mm/year (from 1 December 2017 to 20 November 2018)
*   **Hour** - hour from 0 to 24
*   **Temperature** - celsius temperature
*   **Humidity** - moisture percentage
*   **Wind speed** - in m / s
*   **Rainfall** - precipitation in the form of rain, mm
*   **Snowfall** - precipitation in the form of snow, cm
*   **Seasons** - Winter, Spring, Summer, Autumn 
*   **Holiday** - holiday or non-holiday day
*   **Functioning day** - working or non-working day for bike rental (Yes/No)
*   **Partner 1** - the number of bicycles rented during this hour in the first partner service
*   **Partner 2** - the number of bicycles rented during this hour in the second partner service

```python
bikes  = pd.read_csv('BikesDataImputed.csv', index_col = 'Unnamed: 0')
bikes.head()
```

```python
bikes['Date'] = pd.to_datetime(bikes['Date'], errors='coerce')
bikes['Week'] = bikes['Date'].dt.isocalendar().week
bikes.head()
```

```python
bikes_weeks = bikes.groupby('Week').agg({'Rental Count':np.sum, 
                                         'Temperature':np.mean,
                                         'Humidity':np.mean,
                                         'Wind speed':np.mean})
bikes_weeks.head()
```

```python
sns.pairplot(bikes_weeks)
```

```python
X_mult = bikes_weeks.drop('Rental Count', axis=1)
y = bikes_weeks['Rental Count']
X_mult_constant = sm.add_constant(X_mult)

mod = sm.OLS(y, X_mult_constant)
res_mult = mod.fit()
print(res_mult.summary())
```

## Encoding of variables

```python
bikes.head()
```

### Binary

```python
def precip(x):
  if x>0.0:
    return True
  else:
    return False
```

```python
bikes['rain_check'] = bikes['Rainfall'].apply(precip)

bikes['snow_check'] = bikes['Snowfall'].apply(precip)

bikes['precip_sum'] = bikes['snow_check']+bikes['rain_check']
```

```python
True + True - False # 1 + 1 - 0
```

```python
bikes['Precipitation'] = bikes['precip_sum'].apply(lambda x: 1 if x == True else 0)
bikes.head()
```
